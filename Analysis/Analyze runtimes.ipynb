# Analysis of tool runtimes
Based on AES benchmark test results of 11/13/2018

```python
import pandas
import plotly
plotly.offline.init_notebook_mode(connected=True)
frame = pandas.read_csv("Results/Runtimes_20181113.csv", header = None, names = ["Presynth", "Synth", "Impl", "Run"])
hist_presynth = plotly.graph_objs.Histogram(x = frame["Presynth"], opacity = 0.75, xbins = dict(size = 20), name = "Presynthesis")
hist_synth = plotly.graph_objs.Histogram(x = frame["Synth"], opacity = 0.75, xbins = dict(size = 20), name = "Synthesis")
hist_impl = plotly.graph_objs.Histogram(x = frame["Impl"], opacity = 0.75, xbins = dict(size = 20), name = "Implementation")
hist_run = plotly.graph_objs.Histogram(x = frame["Run"], opacity = 0.75, xbins = dict(size = 20), name = "Run")
layout = plotly.graph_objs.Layout(barmode = 'overlay')
figure = plotly.graph_objs.Figure(data = [hist_presynth, hist_synth, hist_impl, hist_run], layout=layout)
plotly.offline.iplot(figure)
```
