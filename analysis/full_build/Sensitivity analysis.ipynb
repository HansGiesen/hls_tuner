```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
import pandas, numpy
frame = pandas.read_csv("Results/Results_20190205_BNN_random_forest.csv")
frame = frame.loc[~numpy.isinf(frame['run_time'])]

factors = ['CONVOLVERS', 'DATA_MOVER_SHARING', 'DATA_MOVER_CLOCK', 'KERNEL_CLOCK', 'IMPL_STRATEGY']
outputs = ['presynth_time', 'synth_time', 'impl_time', 'run_time']

for output in outputs:
    print(output + ":")
    for factor in factors:    
        mod = ols(output + ' ~ ' + factor, data = frame).fit()
        aov_table = sm.stats.anova_lm(mod, typ=2)
        print aov_table
```
