# Correlate timed-out measurements with parameter values
We print the probability that the state is RTO given that a parameter has a certain value.

```python
sel = frame[frame['state'] == 'RTO']
for param in frame.columns:
    if param == 'id' or param == 'time' or param == 'state':
        continue
    for value in set(frame[param]):
        total = sum(frame[param] == value)
        cnt = sum(sel[param] == value)
        print(param + " = " + str(value) + ": " + str(float(cnt) / total))
```
